```python
import numpy as np

n = 100
x_axis = np.array(list(range(n+1)))
v = np.random.random(n)
```

```python
upsilon_list = []
for _ in range(1000):
    upsilon = np.zeros(n+1)
    for i in range(n):
        x = np.random.choice([-1, 1], p=[0.5, 0.5])
        upsilon[i+1] = upsilon[i] + x * v[i]

    upsilon_list.append(upsilon)
```

```python
upsilon_estimate = [upsilon[-1] ** 2 for upsilon in upsilon_list]
print(np.mean(upsilon_estimate))
```

```python
np.linalg.norm(v) ** 2
```

```python
import matplotlib.pyplot as plt

for i in range(5):  
    plt.plot(x_axis, upsilon_list[i])
plt.plot(x_axis, [np.linalg.norm(v)] * (n+1), '--r', label='||v||')
plt.plot(x_axis, [-np.linalg.norm(v)] * (n+1), '--r')

plt.title('Random Walk')
plt.xlabel('Step')
plt.ylabel('t')
plt.legend()
plt.savefig('random_walk.pdf')
```
